```python
df1=df
df1.drop('Unknown', axis=1, inplace=True)
df1
```

```python
# Select only the rows for age groups
age_group_data = df1.iloc[:-1,]
age_group_data.set_index('Year', inplace=True)
age_group_data
```

```python
# Plot the data as a line chart
plt.figure(figsize=(10, 6))
age_group_data.plot(kind='line', marker='.')
# Add more x-axis values for precision using major and minor ticks
plt.xticks(rotation=45)
plt.gca().xaxis.set_major_locator(plt.MultipleLocator(base=1))
plt.gca().xaxis.set_minor_locator(plt.MultipleLocator(base=1))

# Add more y-axis values for precision using major and minor ticks
plt.yticks(np.arange(0, age_group_data.iloc[:, 2:].max().max() + 1000, 1000))
plt.gca().yaxis.set_minor_locator(plt.MultipleLocator(base=500))

plt.title('Journey of Pregnancies by Age Group (2010-2020) in Sedgwick County', fontname='Arial', fontsize=16, fontweight='bold')
plt.xlabel('Year', fontsize=12, fontname='Arial')
plt.ylabel('Number of Pregnancies', fontsize=12, fontname='Arial')
plt.grid(True)

# Make legend transparent with a white facecolor
legend = plt.legend(title='Age Group', title_fontsize='10', fontsize='10',loc='right')
legend.get_frame().set_facecolor('white')  # Set the background color of the legend to white


# Show the plot
plt.tight_layout()
plt.show()
```

```python

# Filter out the "Total for Selection" row
df_filtered = df[df['Year'] != 'Total for Selection']

# Convert relevant columns to numeric data types
columns_to_convert = ['Year', '10 -- 17', '18 -- 19', '20 -- 24', '25 -- 34', '35 -- 44']
for column in columns_to_convert:
    df_filtered[column] = pd.to_numeric(df_filtered[column], errors='coerce')

# Plotting the data
age_groups = ['10 -- 17', '18 -- 19', '20 -- 24', '25 -- 34', '35 -- 44']
years = df_filtered['Year']
num_age_groups = len(age_groups)
bar_width = 0.15  # Width of each bar
index = np.arange(num_age_groups)  # The x locations for the groups

fig, ax = plt.subplots()

# Plotting the bars for each year and age group
for i, year in enumerate(years):
    data = df_filtered.loc[df_filtered['Year'] == year, age_groups].values.flatten()
    ax.bar(index + bar_width * i, data, bar_width, label=year)

# Format plot elements
ax.set_title('Distribution of Live Births by Age Group (2010-2020) in Sedgwick County')
ax.set_xlabel('Age Group')
ax.set_ylabel('Number of Live Births')
ax.set_xticks(index + bar_width * (num_age_groups - 1) / 2)
ax.set_xticklabels(age_groups)
ax.legend(title='Year')

# Show the plot
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assume you have the data loaded into a DataFrame named 'df'

# Filter the data for a single year (e.g., 2020)
year_data = df[df['Year'] == '2020']

# Remove 'Year' and 'All Ages' columns
age_groups = ['10 -- 17', '18 -- 19', '20 -- 24', '25 -- 34', '35 -- 44']
year_data = year_data[age_groups]

# Convert to numeric
year_data = year_data.apply(pd.to_numeric)

# Calculate the percentage of births for each age group
total_births = year_data.sum().sum()
age_group_percentages = (year_data.sum() / total_births) * 100

# Plotting the pie chart
plt.figure(figsize=(8, 6))
plt.pie(age_group_percentages, labels=age_groups, autopct='%1.1f%%', startangle=140, shadow=True)
plt.title('Distribution of Births by Age Group in 2020', fontsize=16, fontweight='bold')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

# Show the pie chart
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assume you have the data loaded into a DataFrame named 'df'

# Filter out the "Total for Selection" row
df_filtered = df[df['Year'] != 'Total for Selection']

# Convert relevant columns to numeric data types
columns_to_convert = ['Year', '10 -- 17', '18 -- 19', '20 -- 24', '25 -- 34', '35 -- 44']
for column in columns_to_convert:
    df_filtered[column] = pd.to_numeric(df_filtered[column], errors='coerce')

# Calculate the total births for each year
df_filtered['Total'] = df_filtered[age_groups].sum(axis=1)

# Calculate the percentage for each age group in each year
for age_group in age_groups:
    df_filtered[age_group + '_Percentage'] = (df_filtered[age_group] / df_filtered['Total']) * 100

# Plotting the stacked percentage area chart
plt.figure(figsize=(10, 6))
df_filtered.plot(x='Year', y=[age_group + '_Percentage' for age_group in age_groups], kind='area', stacked=True)

# Format plot elements
plt.title('Age Groups Over the Years (Percentage View)', fontsize=16, fontweight='bold')
plt.xlabel('Year', fontsize=12)
plt.ylabel('Percentage', fontsize=12)
plt.legend(title='Age Group', title_fontsize='12', fontsize='10')
plt.grid(False)

# Show the plot
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assume you have the data loaded into a DataFrame named 'df'

# Filter out the "Total for Selection" row
df_filtered = df[df['Year'] != 'Total for Selection']

# Convert relevant columns to numeric data types
age_groups = ['10 -- 17', '18 -- 19', '20 -- 24', '25 -- 34', '35 -- 44']
for column in age_groups:
    df_filtered[column] = pd.to_numeric(df_filtered[column], errors='coerce')

# Create a pivot table for age groups and years
pivot_table = df_filtered.pivot('Year', 'Age', age_groups)

# Plotting the heatmap
plt.figure(figsize=(10, 6))
heatmap = plt.pcolor(pivot_table, cmap='YlGnBu', edgecolors='k', linewidths=0.5)

# Adding color bar
plt.colorbar(heatmap, label='Number of Live Births')

# Format plot elements
plt.title('Distribution of Live Births by Age Group and Year', fontsize=16, fontweight='bold')
plt.xlabel('Age Group', fontsize=12)
plt.ylabel('Year', fontsize=12)

# Adjusting ticks and labels
plt.xticks(range(len(age_groups)), age_groups)
plt.yticks(range(len(pivot_table.index)), pivot_table.index)

# Rotate x-axis labels for better readability
plt.xticks(rotation=45)

# Show the heatmap
plt.tight_layout()
plt.show()
```
