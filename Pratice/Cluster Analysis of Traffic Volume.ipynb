```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from scipy.cluster.hierarchy import linkage, dendrogram

# Load and prepare the dataset
url = "https://raw.githubusercontent.com/VaishnaviKenjale/Road-Traffic-Accidents/refs/heads/main/RTA%20Dataset.csv"
df = pd.read_csv(url)

# Select relevant columns and drop rows with missing values
cols = ['Number_of_vehicles_involved', 'Weather_conditions', 'Light_conditions', 
        'Road_surface_type', 'Type_of_vehicle', 'Cause_of_accident']
data = df[cols].dropna()

# Encode categorical data into numerical format
encoder = LabelEncoder()
encoded_data = data.apply(encoder.fit_transform)

# Standardize the data to have a mean of 0 and standard deviation of 1
scaler = StandardScaler()
scaled_data = scaler.fit_transform(encoded_data)
```

```python
# Apply K-Means clustering to partition the data into 4 clusters
kmeans = KMeans(n_clusters=4, random_state=42, n_init=10)
kmeans_labels = kmeans.fit_predict(scaled_data)

# Reduce dimensionality to 2D for visualization using PCA
pca = PCA(n_components=2)
reduced_data = pca.fit_transform(scaled_data)

# Plot the K-Means clusters
plt.figure(figsize=(10, 6))
plt.scatter(reduced_data[:, 0], reduced_data[:, 1], c=kmeans_labels, cmap='viridis', marker='o')
plt.title("Non-Hierarchical Clustering (K-Means)")
plt.xlabel("Principal Component 1")
plt.ylabel("Principal Component 2")
plt.colorbar(label='Cluster ID')
plt.grid(True)
plt.show()
```

```python
# Perform hierarchical clustering using the 'ward' linkage method
Z = linkage(scaled_data, method='ward')

# Plot the dendrogram to visualize the cluster hierarchy
plt.figure(figsize=(12, 6))
plt.title("Hierarchical Clustering Dendrogram")
plt.xlabel("Data Samples")
plt.ylabel("Distance")
dendrogram(
    Z,
    leaf_rotation=90.,  # rotates the x axis labels
    leaf_font_size=8.,  # font size for the x axis labels
    truncate_mode='lastp',  # show only the last p merged clusters
    p=12,  # show only the last 12 merged clusters
    show_contracted=True,  # to get a tidy plot
)
plt.show()
```

```python
# Importing all the stuff I need
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from scipy.cluster.hierarchy import linkage, dendrogram

# ---- Loading the dataset ----
# Got this dataset from GitHub (it’s about road traffic accidents)
url = "https://raw.githubusercontent.com/VaishnaviKenjale/Road-Traffic-Accidents/refs/heads/main/RTA%20Dataset.csv"
df = pd.read_csv(url)

# Just checking what the data looks like
# print(df.head())   # (I commented this out to avoid cluttering output)

# ---- Selecting columns that I think are important ----
cols_needed = [
    'Number_of_vehicles_involved', 
    'Weather_conditions', 
    'Light_conditions', 
    'Road_surface_type', 
    'Type_of_vehicle', 
    'Cause_of_accident'
]

data = df[cols_needed].dropna()   # removing rows with missing stuff

# ---- Encoding the categorical columns ----
# using LabelEncoder because there are string categories
encoder = LabelEncoder()
encoded_df = data.apply(encoder.fit_transform)

# ---- Scaling the data ----
# This step is to make sure all features are on the same scale
scaler = StandardScaler()
scaled_values = scaler.fit_transform(encoded_df)

# just converting it back to a DataFrame for easier inspection
scaled_df = pd.DataFrame(scaled_values, columns=cols_needed)

# ---- K-Means Clustering ----
# I’ll try 4 clusters (random choice for now)
kmeans_model = KMeans(n_clusters=4, random_state=42, n_init=10)
kmeans_labels = kmeans_model.fit_predict(scaled_df)

# ---- PCA (to visualize clusters in 2D) ----
pca = PCA(n_components=2)
reduced = pca.fit_transform(scaled_df)

# plotting the clusters
plt.figure(figsize=(9, 6))
plt.scatter(reduced[:, 0], reduced[:, 1], c=kmeans_labels, cmap='plasma', s=50)
plt.title("K-Means Clustering (after PCA)")
plt.xlabel("PC1")
plt.ylabel("PC2")
plt.colorbar(label='Cluster Group')
plt.grid(True)
plt.show()

# ---- Hierarchical Clustering ----
# using 'ward' linkage as it usually gives decent results
Z = linkage(scaled_df, method='ward')

# Drawing dendrogram (kinda messy for large data)
plt.figure(figsize=(12, 6))
plt.title("Hierarchical Clustering Dendrogram")
plt.xlabel("Samples")
plt.ylabel("Distance")
dendrogram(
    Z,
    leaf_rotation=90.,
    leaf_font_size=8.,
    truncate_mode='lastp',  # showing last 12 clusters for readability
    p=12,
    show_contracted=True
)
plt.show()

# ---- Note to self ----
# Could try DBSCAN or AgglomerativeClustering next time for comparison.
# Also maybe find out what each cluster actually means...
```
